```python
import numpy as np
import matplotlib.pyplot as plt

# 진동 파라미터 설정
mass = 1.0  # 질량 (kg)
spring_constant = 5.0  # 탄성 계수 (N/m)
damping_coefficient = 0.3  # 감쇄 계수 (N*s/m)

# 초기 조건
initial_position = 1.0  # 초기 위치 (m)
initial_velocity = 0.0  # 초기 속도 (m/s)

# 시간 설정
time = np.linspace(0, 30, 1000)  # 0부터 10초까지 1000개의 시간 스텝

# 진동 방정식
omega0 = np.sqrt(spring_constant / mass)  # 고유진동수
omega_d = np.sqrt(omega0**2 - (damping_coefficient / (2 * mass))**2)  # 감쇄진동수
position = (
    initial_position * np.exp((-damping_coefficient / (2 * mass)) * time)
    * np.cos(omega_d * time)
)

# 그래프 그리기
plt.figure(figsize=(8, 6))

# 배경색을 검정색으로 설정
plt.style.use('dark_background')

plt.plot(time, position, label='Position', color='white')
plt.title('Damped Oscillation')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.legend()
plt.grid(True)
plt.show()
```
