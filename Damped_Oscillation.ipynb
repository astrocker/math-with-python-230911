```python
import numpy as np
import matplotlib.pyplot as plt

#   
mass = 1.0  #  (kg)
spring_constant = 5.0  #   (N/m)
damping_coefficient = 0.3  #   (N*s/m)

#  
initial_position = 1.0  #   (m)
initial_velocity = 0.0  #   (m/s)

#  
time = np.linspace(0, 30, 1000)  # 0 10 1000  

#  
omega0 = np.sqrt(spring_constant / mass)  # 
omega_d = np.sqrt(omega0**2 - (damping_coefficient / (2 * mass))**2)  # 
position = (
    initial_position * np.exp((-damping_coefficient / (2 * mass)) * time)
    * np.cos(omega_d * time)
)

#  
plt.figure(figsize=(8, 6))

#   
plt.style.use('dark_background')

plt.plot(time, position, label='Position', color='white')
plt.title('Damped Oscillation')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.legend()
plt.grid(True)
plt.show()
```
