```python
import numpy as np
import matplotlib.pyplot as plt

# 원의 반지름 입력
radius = float(input("Enter the radius of the circle: "))

# 원의 넓이 계산
area = np.pi * radius ** 2
print(f"The area of the circle is {area:.2f} square units.")

# 그래프 그리기
theta = np.linspace(0, 2 * np.pi, 100)
x = radius * np.cos(theta)
y = radius * np.sin(theta)

plt.figure(figsize=(6, 6))

# 배경색을 검정색으로 설정
plt.style.use('dark_background')

plt.plot(x, y, color='white')  # 그래프 색상을 흰색으로 설정
plt.title("Graph of a Circle")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 원의 중심 (a, b)와 반지름 r 설정
a = 2.0
b = 3.0
r = 2.5

# x 값 배열 생성
x_values = np.linspace(a - r, a + r, 200)  # 원의 반지름을 고려하여 x 범위 설정

# 해당 x 값에 대한 위쪽 반원의 y 값 계산
y_values_upper = b + np.sqrt(r**2 - (x_values - a)**2)

# 해당 x 값에 대한 아래쪽 반원의 y 값 계산
y_values_lower = b - np.sqrt(r**2 - (x_values - a)**2)

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.style.use('dark_background')
plt.plot(x_values, y_values_upper, color='white')  # 위쪽 반원 그래프
plt.plot(x_values, y_values_lower, color='white')  # 아래쪽 반원 그래프
plt.title("Graph of a Circle")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
