```python
import numpy as np
import matplotlib.pyplot as plt

# 시간 범위 생성
t = np.linspace(0, 2 * np.pi, 1000)

# 푸리에 급수 항의 개수 선택
num_terms = 10

# 푸리에 급수 계산
def fourier_series(t, num_terms):
    series = np.zeros_like(t)
    for n in range(1, num_terms + 1):
        term = (4 / (np.pi * n)) * np.sin((2 * np.pi * n * t) / (2 * np.pi))
        series += term
    return series

plt.figure(figsize=(8, 6))

# 배경색을 검정색으로 설정
plt.style.use('dark_background')

# 푸리에 급수 계산
series = fourier_series(t, num_terms)

# 원래의 사인 파형 그리기
original_signal = np.sin(t)

plt.plot(t, original_signal, label='Original Signal', color='white')
plt.plot(t, series, label=f'{num_terms} Terms of Fourier Series', color='red')

plt.title(f'Approximation of Sine Wave using {num_terms} Terms of Fourier Series')
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 진동 방정식의 파라미터 설정
m = 1.0  # 진동체의 질량 (mass)
c = 0.2  # 감쇄 계수 (damping coefficient)
k = 1.0  # 탄성 계수 (spring constant)

# 초기 조건 설정
x0 = 1.0  # 초기 위치
v0 = 0.0  # 초기 속도

# 시간 관련 설정
t_max = 10.0  # 시뮬레이션 시간
dt = 0.01    # 시간 간격

# 시간 배열 생성
t = np.arange(0, t_max, dt)

# 위치와 속도 초기화
x = np.zeros_like(t)
v = np.zeros_like(t)
x[0] = x0
v[0] = v0

# 진동 방정식을 해결하는 단계적인 시뮬레이션
for i in range(1, len(t)):
    # 진동 방정식 업데이트
    F_spring = -k * x[i - 1]  # 탄성력
    F_damping = -c * v[i - 1]  # 감쇄력
    F_total = F_spring + F_damping
    a = F_total / m  # 가속도
    v[i] = v[i - 1] + a * dt
    x[i] = x[i - 1] + v[i] * dt

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.style.use('dark_background')
plt.plot(t, x, color='white', label='Displacement (x)')
plt.plot(t, v, color='red', label='Velocity (v)')
plt.title("Vibratory Motion")
plt.xlabel("Time (s)")
plt.ylabel("Displacement / Velocity")
plt.legend()
plt.grid(True)
plt.show()
```
