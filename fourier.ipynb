```python
import numpy as np
import matplotlib.pyplot as plt

# 시간 범위 생성
t = np.linspace(0, 2 * np.pi, 1000)

# 푸리에 급수 항의 개수 선택
num_terms = 10

# 푸리에 급수 계산
def fourier_series(t, num_terms):
    series = np.zeros_like(t)
    for n in range(1, num_terms + 1):
        term = (4 / (np.pi * n)) * np.sin((2 * np.pi * n * t) / (2 * np.pi))
        series += term
    return series

plt.figure(figsize=(8, 6))

# 배경색을 검정색으로 설정
plt.style.use('dark_background')

# 푸리에 급수 계산
series = fourier_series(t, num_terms)

# 원래의 사인 파형 그리기
original_signal = np.sin(t)

plt.plot(t, original_signal, label='Original Signal', color='white')
plt.plot(t, series, label=f'{num_terms} Terms of Fourier Series', color='red')

plt.title(f'Approximation of Sine Wave using {num_terms} Terms of Fourier Series')
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.legend()
plt.grid(True)
plt.show()
```
